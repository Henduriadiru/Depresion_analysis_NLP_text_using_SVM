```python
from sklearn.feature_extraction.text import ENGLISH_STOP_WORDS
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer
import nltk
from collections import Counter
import re
```

```python
nltk.download('punkt')
nltk.download('wordnet')
```

```python
lemmatizer = WordNetLemmatizer()

df_combined["tweet"] = df_combined["tweet"].apply(lambda x : re.sub(r'(@\w+|\d+|[^\w\s])', '', x).lower().strip())
df_depres_filter["tweet"] = df_depres_filter["tweet"].apply(lambda x : re.sub(r'(@\w+|\d+|[^\w\s])', '', x).lower().strip())

def clean_text(text):
    token_cleaned = [lemmatizer.lemmatize(word) for word in word_tokenize(text) if word not in ENGLISH_STOP_WORDS and word.isalpha()]
    return token_cleaned

df_combined["cleaned_tweet"] = df_combined['tweet'].apply(clean_text)
df_depres_filter["cleaned_tweet"] = df_combined['tweet'].apply(clean_text)
df_combined
```

```python
label_counts = df_combined['label'].value_counts()

fig, (axs1, axs2) = plt.subplots(1, 2, figsize=(10,5))

axs1.pie(label_counts, labels=["Positive", "Negative"], autopct="%1.1f%%", startangle=90, colors=["red", "blue"])
axs1.set_title("Sentimen Depression Distribution Pie")

axs2.bar(label_counts.index, label_counts.values, color=["red", "blue"])
axs2.set_title('Sentiment Distribution (Histogram)')
axs2.set_xlabel('Labels')
axs2.set_ylabel('Count')
ax2.set_xticks([0,1])

plt.tight_layout()
plt.show()
```

```python
# Count most common word in depresion data
all_words = [word for tokens in df_depres_filter['cleaned_tweet'] for word in tokens]

word_counts = Counter(all_words)

most_common_word_use = word_counts.most_common(10)

words, freq = zip(*most_common_word_use)

plt.figure(figsize=(10,6))

plt.bar(words, freq, color= "skyblue")
plt.title("Top 10 Word Common Use")
plt.xlabel("Words")
plt.ylabel("Frequent")
plt.xticks(rotation=45)

plt.show()
```

```python
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
from sklearn.manifold import TSNE

# Menggunakan CountVectorizer untuk mendapatkan bigram
cvectorizer = CountVectorizer(ngram_range=(2, 2), stop_words="english")

X = cvectorizer.fit_transform(df_depres_filter["tweet"])
bigrams = X.sum(axis=0).A1 # Menghitung jumlah bigram

bigrams_dict = dict(zip(cvectorizer.get_feature_names_out(), bigrams)) # Menyusun bigram dan frekuensinya dalam bentuk dictionary

sorted_bigrams = sorted(bigrams_dict.items(), key= lambda x : x[1], reverse=True) # Menyusun bigram berdasarkan frekuensi

bigrams, freq = zip(*sorted_bigrams[:10])

plt.figure(figsize=(10, 5))
plt.bar(bigrams, freq, color='lightcoral')
plt.title('Top 10 Bigrams')
plt.xlabel('Bigram')
plt.ylabel('Frequency')
plt.xticks(rotation=45)
plt.show()
```

```python
from gensim.models import Word2Vec
from sklearn.decomposition import PCA
vectorizer = TfidfVectorizer(stop_words="english")
X = vectorizer.fit_transform(df_depres_filter['tweet'])

model = Word2Vec(df_combined["cleaned_tweet"], vector_size=100, window=5, min_count=1, workers=4)

def text_to_vector(clean_text, model):
    vectors = [model.wv[token] for token in clean_text if token in model.wv]
    if len(vectors)==0:
        return np.zeros(model.vector_size)
    return np.mean(vectors, axis=0)

vector_wv = np.array([text_to_vector(clean_text, model) for clean_text in df_combined['cleaned_tweet']])

pca = PCA(n_components=50)
X_pca = pca.fit_transform(vector_wv)

plt.figure(figsize=(10,5))
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=df_combined['label'], cmap="viridis")
plt.title("PCA visualisation of Depression Text")
plt.xlabel("PCA component 1")
plt.ylabel("PCA component 2")
plt.colorbar(label="label")
plt.show()

tsne = TSNE(n_components=2, random_state=42)
tsne_result = tsne.fit_transform(X_pca)

plt.figure(figsize=(10,5))
plt.scatter(tsne_result[:, 0], tsne_result[:, 1], c=df_combined['label'], cmap="viridis")
plt.title("t-SNE visualisation of Depression Text")
plt.xlabel("t-SNE component 1")
plt.ylabel("t-SNE component 2")
plt.colorbar(label="label")
plt.show()
```

```python
from sklearn.svm import SVC
from sklearn.model_selection import RandomizedSearchCV, train_test_split
from sklearn.metrics import classification_report
from sklearn.preprocessing import StandardScaler
from tqdm import tqdm

def clasification_using_svm(X_train, X_test, y_train, y_test):

    svm = SVC()

    param = {
        "C" : [0.1, 1.0, 10.0],
        "kernel" : ["linear", "rbf"],
        "gamma" : ["scale", "auto"]
    }
    gridsearch = RandomizedSearchCV(svm, param_distributions= param, n_iter=10 ,cv=5, scoring= "accuracy", random_state=42, n_jobs=-1, verbose=0)

    n_iter = 10
    with tqdm(total=10, desc="RandomizedSearchCV Progress") as pbar:
    # Fit the model and update progress bar during each iteration
        gridsearch.fit(X_train, y_train)
        pbar.update(10)

    print("Best Parameters :", gridsearch.best_params_)
    y_pred = gridsearch.predict(X_test)
    print(classification_report(y_test, y_pred))
    results = pd.DataFrame(gridsearch.cv_results_)
    display(results)

    #plot_svm_decision_boundary(X_train, y_train, gridsearch.best_estimator_)

    return gridsearch

```

```python
def plot_svm_decision_boundary(X, y, model):

    h = .02
    x_min, x_max = X[:, 0].min() -1, X[:, 0].max()+1
    y_min, y_max = X[:, 1].min() -1, X[:, 1].max()+1
    xx, yy = np.meshgrid(np.linspace(x_min, x_max, 100), np.linspace(y_min, y_max, 100))

    Z = model.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)

    plt.figure(figsize=(10,5))
    plt.contour(xx, yy, Z, alpha=0.8, cmap="coolwarm")

    plt.scatter(X[:, 0], X[:, 1], c=y, cmap='coolwarm', edgecolors='k', s=50)
    plt.title("SVM Decision Boundary with RBF Kernel")
    plt.xlabel("PCA Component 1")
    plt.ylabel("PCA Component 2")
    plt.colorbar(label="Label")
    plt.show()
    
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_pca, df_combined["label"], test_size=0.2 ,random_state=42)
clasification_using_svm(X_train, X_test, y_train, y_test)
```
